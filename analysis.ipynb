```python

def plot_heat_map(spotify_corr):   
    mask = np.triu(np.ones_like(spotify_corr, dtype=bool))
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    plt.figure(figsize=(12, 12))
    sns.heatmap(spotify_corr, mask=mask, cmap=cmap, vmax=1, vmin=-1, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .75}, annot=True, fmt=".2f", annot_kws={"size": 10})
    cbar = plt.gcf().axes[-1]
    cbar.set_ylabel('Correlation Coefficient', rotation=270, labelpad=20, fontsize=12)
    plt.title('Correlation Matrix', fontsize=16)
    plt.xticks(fontsize=12)
    plt.yticks(fontsize=12)
    plt.tight_layout()
    plt.show()


def plot_features_heatmap(data):
    features=['danceability', 'energy', \
       'loudness', 'speechiness', \
        'acousticness', 'instrumentalness', \
       'liveness', 'valence', 'tempo']
    data=data.drop('Unnamed: 0',axis=1)
    numerical_data = (data.filter(items=features)).fillna(0)
    plot_heat_map(spotify_corr=numerical_data.corr())
    return numerical_data

numerical_spotify_data=plot_features_heatmap(data=data)
```

```python
    
def scaling_numerical_data(numerical_data):

    from sklearn.preprocessing import MinMaxScaler
    scaler = MinMaxScaler()
    scaled_array = scaler.fit_transform(numerical_data)
    scaled_df=pd.DataFrame(scaled_array,columns=numerical_data.columns)   
    return scaled_df


scaled_df=scaling_numerical_data(numerical_data=numerical_spotify_data)
scaled_df
```

```python
import torch
```

```python
def turn_on_GPU():
    device=torch.device('cuda' if torch.cuda.is_available() else 'cpu' )
    return device

device=turn_on_GPU()
```

```python
scaled_df_tensor=torch.tensor(scaled_df_only_values,dtype= torch.float32).to(device)
```

```python
scaled_df_tensor.shape
```

```python
import torch.nn as nn
import torch.optim as optim
```

```python
class Autoencoder(nn.Module):
    def __init__(self,input_dim,encoding_dim):
        super(Autoencoder,self).__init__()
        self.encoder=nn.Sequential(
            nn.Linear(input_dim,encoding_dim),
            nn.ReLU()
        )
        self.decoder=nn.Sequential(
            nn.Linear(encoding_dim,input_dim),
            nn.Sigmoid()
        )
    def forward(self,x):
        encoded=self.encoder(x)
        decoded=self.decoder(encoded)
        return decoded

input_dim=scaled_df_tensor.shape[1]
print(input_dim)
encoding_dim=6
model=Autoencoder(input_dim=input_dim,encoding_dim=encoding_dim).to(device=device)

```

```python
criterion=nn.MSELoss()
optimizer=optim.Adam(model.parameters(),lr=0.001)
```

```python
num_epochs = 100
batch_size = 256

dataset=torch.utils.data.TensorDataset(scaled_df_tensor,scaled_df_tensor)
dataloader=torch.utils.data.DataLoader(dataset,batch_size=batch_size,shuffle=True)
```

```python
model.train()
for epoch in range(num_epochs):
    for data in dataloader:
        inputs, _=data
        inputs=inputs.to(device=device)
        
        outputs=model(inputs)
        loss=criterion(outputs,inputs)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    print(f'EPOCH: [{epoch+1}/{num_epochs}], LOSS={loss.item():.4f}')
```

```python
model.eval()
with torch.no_grad():
    song_embeddings=model.encoder(scaled_df_tensor).cpu().numpy()
```

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE

# Assuming song_embeddings is a PyTorch tensor, convert it to a NumPy array
song_embeddings_np = song_embeddings

# Fit KMeans clustering
num_clusters = 10
kmeans = KMeans(n_clusters=num_clusters, random_state=0)
clusters = kmeans.fit_predict(song_embeddings_np)

# Reduce dimensionality to 2D using t-SNE
tsne = TSNE(n_components=2, random_state=42)
embeddings_2d = tsne.fit_transform(song_embeddings_np)

# Plot the clusters
plt.figure(figsize=(10, 7))
for cluster in range(num_clusters):
    cluster_points = embeddings_2d[clusters == cluster]
    plt.scatter(cluster_points[:, 0], cluster_points[:, 1], label=f'Cluster {cluster}')

plt.title('Clusters of Songs in 2D Space (t-SNE)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.legend()
plt.show()
```
